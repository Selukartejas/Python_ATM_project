```python
class welcome:
    def print_welcome(self,obj):
        print("                       # WELCOME TO BANK OF INDIA #")
        print("                     --------------------------------")
        obj.get_authenticate().print_authenticate(obj)
        
class authenticate:
    inp = 3002
    def print_authenticate(self,obj):
        pin = int(input("Enter your pin:"))
        if(pin==self.inp):
            obj.get_menu().print_menu(obj)
        elif(pin!=self.inp):
            print("please enter the correct pin")
            pin = int(input("Enter your pin:"))
            if(pin==self.inp):
                obj.get_menu().print_menu(obj)
            elif(pin!=self.inp):
                print("please enter the correct pin")
                pin = int(input("Enter your pin:"))
                if(pin==self.inp):
                    obj.get_menu().print_menu(obj)
                elif(pin!=self.inp):
                    print("------BLOCK-----")
                    
    def print_pchange(self,obj):
        print("------------------4.Pin Change:---------------")
        op = int(input("Enter your old pin:"))
        if(op==self.inp):
            np = int(input("Enter your new pin:"))
            self.inp = np
            print("Your new pin is generated.")
            obj.get_authenticate().print_authenticate(obj)
        else:
            print("You are enter wrong pin!")
            obj.get_authenticate().print_pchange(obj)
            
            
            
class menu:
    def print_repeat(self,obj):
        ans = input("You are enter another option(yes/no):")
        if(ans=="yes"):
            return True
        elif(ans=="no"):
            return False
            
    def print_menu(self,obj):
        flag = True
        while flag:
            print("\n# Select your option:")
            print("-----------------------")
            print(" 1.Check Balance")
            print(" 2.Withdraw")
            print(" 3.Deposit")
            print(" 4.Pin change")
            print(" 5.Exit")
            opt = int(input("\nEnter your option:"))
            if(opt==1):
                obj.get_balance().print_balance(obj)
                flag = obj.get_menu().print_repeat(obj)
            elif(opt==2):
                obj.get_withdraw().print_withdraw(obj)
                flag = obj.get_menu().print_repeat(obj)
            elif(opt==3):
                obj.get_deposite().print_deposite(obj)
                flag = obj.get_menu().print_repeat(obj)
            elif(opt==4):
                obj.get_authenticate().print_pchange(obj)
                flag = obj.get_menu().print_repeat(obj)
            elif(opt==5):
                flag= False
                print("------------------5.Exit:----------------------")
                print("               Thanking You!")
                print("                  B.O.I")
#                 flag= False
            else:
                print("Please,enter your corrct option!")
                obj.get_menu().print_menu(obj)
            
            
class balance:
    def print_balance(self,obj):
        print("\n------------------1.Balance:-------------------")
        total = obj.get_ledger().print_balance(obj)
        print("Your Balance:",total)

class ledger:
    balance = 10000
    def print_balance(self,obj):
        return self.balance
    def print_withdraw_amt(self,obj,w_amt):
        self.balance = self.balance - w_amt
    def print_deposite_amt(self,obj,d_amt):
        self.balance = self.balance + d_amt
        
class withdraw:
    def print_withdraw(self,obj):
        print("\n------------------2.Withdraw:------------------")
        w_amt = int(input("Enter your withdraw amount:"))
        obj.get_ledger().print_withdraw_amt(obj,w_amt)
        total = obj.get_ledger().print_balance(obj)
        if(total>=w_amt):
            print("You are remaining amount is",total,"Rs")
        else:
            print("You have not sufficent balance.your balance is",total)

class deposite:  
    def print_deposite(self,obj):
        print("\n------------------3.Deposit:-------------------")
        d_amt = int(input("Enter your deposite amount:"))
        total = obj.get_ledger().print_deposite_amt(obj,d_amt)
        total = obj.get_ledger().print_balance(obj)
        print("Current Balnce :",total)
        
class objfact:
    w = welcome()
    a = authenticate()
    m = menu()
    b = balance()
    v = withdraw()
    d = deposite()
    l = ledger()
    def get_welcome(self):
        return self.w
    def get_authenticate(self):
        return self.a
    def get_menu(self):
        return self.m
    def get_balance(self):
        return self.b
    def get_withdraw(self):
        return self.v
    def get_deposite(self):
        return self.d
    def get_ledger(self):
        return self.l
    
obj = objfact()
s = obj.get_welcome().print_welcome(obj)
```
